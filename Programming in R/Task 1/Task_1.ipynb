# Introduction

Collab is the interactive notebook of Google (what Jupyter Notebook is for Python). It permits writing and running code through a browser, using a Google GPU without additional cost. Similarly to Google Drive and Docs, it makes easier to share and edit code between several users at the same time.

While Colab is primarily utilized for Python coding, it can also be employed for R programming. In the next section, we are going to explain how to integrate R on Google Colab.


# Running R in Colab



As has been mentioned before, Colab was initially designed for coding in the Python language, although later on the possibility of running R code was added to the platform.
In order to do so, there are two possibilities, both of which will be explained.

The first option allows for the implementation of both Python and R code within the same notebook. To do so, one must make use of the rpy2 package within a Python runtime-type Colab notebook, that is, a standard Colab notebook. The first step is to execute the following command within a code cell:


```
%load_ext rpy2.ipython
```



This starts *rmagic* within the notebook. \\
Once that code has been executed, every time the user might want to execute some R code snippet within the notebook, the machine has to be explicitly told in any of two options. \\
If the user wants to run a whole cell as R code, the command ```%%R``` will have to be placed at the very start of said cell:

```
%%R
x <- seq(0, 2*pi, length.out=50)
```
This is known as **cell magic**.

On the other hand, the user might want to simply run one or several lines of code as opposed to the whole cell, so that both Python and R can be combined within the same cell. For this, **line magic** is more appropiate. For example, the code shown below would copy the object *x* generated within an R-code cell to a Python object *x*:
```
x = %R x
```

The first step is to open a Colab notebook. This notebook, with file expansion .ipynb, will be prepared to execute Python code, just as any Jupyter notebook would. Altering the notebook to allow for the execution of exclusively Python code is fairly simple, doing a couple easy steps.
